


Write the regular expression pattern for matching email addresses in Python?


```python
import re

text = 'rahul123@gmail.com'
pattern = re.compile(r'^\w+\d+\@+\w+\.\w+$')

m  = re.search(pattern, text)
m.group()
```

Write a piece of Code for removing whitespace from the beginning and end of a string in
Python using regular expressions?



```python
import re

text = "   rahul123@gmail.com           "
pattern = re.compile(r'^\s+|\s+$')

s = re.sub(pattern, '', text)
s
```

How do you extract data from an HTML page using BeautifulSoup in Python?


```python
from bs4 import BeautifulSoup

URL = 'https://www.wikipedia.org/'
fetch = requests.get(URL)
soup = BeautifulSoup(fetch.text, parser= 'html.parser')
soup
```
